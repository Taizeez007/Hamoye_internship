```python
import pandas as pd
import numpy as np
import requests
import io

url="https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"

fuel_dt =requests.get(url).content#use request command 'get' to get data from online repository 

fuel_data=pd.read_csv(io.StringIO(fuel_dt.decode('utf-8')))#checking another way to get data through url

fuel_skew=fuel_data['fuel_qty_burned'].skew()#getting skewness of quantity of fuel burned
print(fuel_skew)

fuel_kurtosis=fuel_data['fuel_qty_burned'].kurtosis()#finding kurtosis of quantity of fuel burned
print(fuel_kurtosis)

fuel_percentile = np.percentile(fuel_data['fuel_mmbtu_per_unit'],75,axis=None,out=None)#75 percentile of measuer of energy per unit
print(fuel_percentile)

fuel_per_unit_std =np.std(fuel_data['fuel_mmbtu_per_unit'])#standard deviation of fuel_mmbtu_per_unit
print(fuel_per_unit_std)

feature_1=fuel_data['fuel_cost_per_unit_burned']
feature_list=list['fuel_mmbtu_per_unit','fuel_qty_burned','fuel_cost_per_unit_delivered','fuel_cost_per_mmbtu','report_year','utility_id_ferc1']
correlation=feature_1.corrwith(fuel_data, axis=0)
```
